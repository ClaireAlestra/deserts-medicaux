```python
#Base de données mergée avec etab et entities
merge_df = etab_df.merge(entities_df, on='siren', how='outer', indicator=True)
```

```python
#Filtrer les etab en 86 ou les entités en 86 (Filtrage des professionels de santé)
merged_df = merge_df[merge_df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE) | merge_df.activiteprincipaleunitelegale.str.startswith(ACTIVITY_CODE)]
```

```python
merged_df['_merge'].value_counts()
```

```python
#Analyser quels sont les types d'écart entre les activités etab et les entités
#Tableau croisé activité etab et entité
pd.crosstab(merged_df.activiteprincipaleunitelegale.str.startswith(ACTIVITY_CODE), merged_df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE)) 
```

Donc nous avons 7954 unité légales classées comme professionels de santé (86) alors que les établissements correspondants ne le sont pas.

Nous avons également 21595 établissements classés comme professionels de santé (86) alors que les unités légales correspondantes ne le sont pas.

Nous avons 552149 correspondances (établissements et unités légales tous deux classés comme professionels de santé: 86).

```python
ps_df = (merged_df.loc[merged_df.activiteprincipaleunitelegale.str.startswith("86.") |
                      merged_df.activiteprincipaleetablissement.str.startswith("86.")]
         .assign(etab_ape_l1 = lambda df: df.activiteprincipaleetablissement.str.slice(0, 2))
         .assign(el_ape_l1 = lambda x: x.activiteprincipaleunitelegale.str.slice(0, 2))
         .assign(cross_l1 = lambda df: df.el_ape_l1 + '_' + df.etab_ape_l1)
        )
```

```python
ps_df.cross_l1.value_counts()
```

Donc on a juste besoin des 4 premières catégories pour avoir la majorité de la population.

## Préparation  de la table des codes NAF

```python
#coder le code d'activité 86 (règle de choix): 86_86 -> etab 86_x ou y -> 86
ps_df.loc[ps_df.cross_l1 == "86_86", 'code_activite'] = ps_df.activiteprincipaleetablissement
ps_df.loc[ps_df.el_ape_l1 != "86", 'code_activite'] = ps_df.activiteprincipaleetablissement
ps_df.loc[~ps_df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE), 'code_activite'] = ps_df.activiteprincipaleunitelegale
```

```python
#autre manière de faire:
ps_df['code_activite'] = ps_df.apply(lambda x: x.activiteprincipaleetablissement 
                                     if x.etab_ape_l1 == "86"
                                     else x.activiteprincipaleunitelegale, 
                                     axis=1
                                     )
```

```python
#troisème version:
ps_df['code_activite'] = ps_df.activiteprincipaleetablissement
ps_df['code_activite'] = ps_df.where(ps_df.etab_ape_l1 != "86", 
                                     ps_df.activiteprincipaleunitelegale,
                                     axis= 0
                                     )
```

```python
#quatrième version:
import numpy as np

ps_df['code_activite'] = np.where(ps_df.etab_ape_l1 == '86',
                                  ps_df.activiteprincipaleetablissement,
                                  ps_df.activiteprincipaleunitelegale
                                  )
```

## Ajout du type de PS et sauvegarde

```python
#Base de données mergée avec naf, puis tranposée, puis enregistrée en csv
(ps_df
 .merge(naf_df, how='inner', left_on='code_activite', right_on='NAF')
 [['codecommuneetablissement', 'Catégorie']]
 .dropna(subset=['codecommuneetablissement'])
 .assign(count = 1)
 .groupby(['codecommuneetablissement', 'Catégorie'])
 .sum()
 .reset_index()
 .pivot(index='codecommuneetablissement', columns='Catégorie', values='count')
 .fillna(0)
).to_csv(os.path.join(PROCESSED_DIR, PS_CITIES_FILENAME), sep=';')
```
